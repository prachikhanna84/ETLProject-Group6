```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
city_file = "../Resources/city_attributes.csv"
city_data_df = pd.read_csv(city_file)
city_data_df.head()
```

```python
humidity_file = "../Resources/humidity.csv"
humidity_data_df = pd.read_csv(humidity_file)
humidity_data_df=pd.melt(humidity_data_df, id_vars=['datetime'])

humidity_data_df.head()
```

```python
wind_speed_file = "../Resources/wind_speed.csv"
wind_speed_data_df = pd.read_csv(wind_speed_file)
wind_speed_data_df=pd.melt(wind_speed_data_df, id_vars=['datetime'])
wind_speed_data_df.head()
```

```python
wind_direction_file = "../Resources/wind_direction.csv"
wind_direction_data_df = pd.read_csv(wind_direction_file)
wind_direction_data_df=pd.melt(wind_direction_data_df, id_vars=['datetime'])
wind_direction_data_df.head()
```

```python
#Load and Read Files
temp_file = "../Resources/temperature.csv"
temp_data_df = pd.read_csv(temp_file)
temp_data_df=pd.melt(temp_data_df, id_vars=['datetime'])
temp_data_df.head()
```

```python
#Load and Read Files
pres_file = "../Resources/pressure.csv"
pres_data_df = pd.read_csv(pres_file)
pres_data_df=pd.melt(pres_data_df, id_vars=['datetime'])

pres_data_df.head()
```

```python
#Load and Read Files
weather_description_file = "../Resources/weather_description.csv"
weather_description_data_df = pd.read_csv(weather_description_file)
weather_description_data_df=pd.melt(weather_description_data_df, id_vars=['datetime'])
weather_description_data_df.head()
```

```python
 # Create a filtered dataframe from specific columns
city_mapping_cols = ['City', 'Country', 'Latitude','Longitude']
city_transformed= city_data_df[city_mapping_cols].copy()

# Rename the column headers
city_transformed = city_transformed.rename(columns={'City':'city', 'Country':'country', 'Latitude':'latitude','Longitude':'longitude'})

# city_transformed.set_index("city",inplace=True)
city_transformed.head()
```

```python
 # Create a filtered dataframe from specific columns
humidity_mapping_cols = ['datetime', 'variable','value']
humidity_transformed= humidity_data_df[humidity_mapping_cols].copy()
# Rename the column headers
humidity_transformed = humidity_transformed.rename(columns={'datetime':'datetime', 'variable':'city','value':'humidity'})

# humidity_transformed.set_index("humidity",inplace=True)
humidity_transformed.head()
```

```python
  # Create a filtered dataframe from specific columns
wind_speed_mapping_cols = ['datetime', 'variable','value']
wind_speed_transformed= wind_speed_data_df[wind_speed_mapping_cols].copy()
# Rename the column headers
wind_speed_transformed = wind_speed_transformed.rename(columns={'datetime':'datetime', 'variable':'city','value':'wind_speed'})

# wind_speed_transformed.set_index("wind_speed",inplace=True)
wind_speed_transformed.head()
```

```python
  # Create a filtered dataframe from specific columns
wind_direction_mapping_cols = ['datetime', 'variable','value']
wind_direction_transformed= wind_direction_data_df[wind_direction_mapping_cols].copy()
# Rename the column headers
wind_direction_transformed = wind_direction_transformed.rename(columns={'datetime':'datetime', 'variable':'city','value':'wind_direction'})


bins = [0, 90, 180, 270, 360]

# Create the names for the four bins
group_names = ["NE", "SE", "SW", "NW"]

wind_direction_transformed["wind_direction"] = pd.cut(wind_direction_transformed["wind_direction"], bins, labels=group_names)
wind_direction_transformed.head()


# wind_direction_transformed.set_index("wind_direction",inplace=True)
wind_direction_transformed.head()
```

```python
#Map Column Names
temp_df = temp_data_df[['datetime', 'variable','value']]
temp_col_mapping = { 
                  "datetime": "datetime",
                  "variable": "city",
                  "value": "temperature"}
temp_df = temp_df.rename(columns=temp_col_mapping)
temp_df.head()
```

```python
#Convert Kalvin to Fahrenheit
temperature_f = (temp_df["temperature"]-273.15)*9/5 + 32
temp_df["temperature"] = temperature_f
temp_df.head()
```

```python
  # Create a filtered dataframe from specific columns
pressure_mapping_cols = ['datetime', 'variable','value']
pressure_transformed= pres_data_df[pressure_mapping_cols].copy()
# Rename the column headers
pressure_transformed = pressure_transformed.rename(columns={'datetime':'datetime', 'variable':'city','value':'pressure'})

pressure_transformed.head()
```

```python
  # Create a filtered dataframe from specific columns
weather_description_mapping_cols = ['datetime', 'variable','value']
weather_description_transformed= weather_description_data_df[weather_description_mapping_cols].copy()
# Rename the column headers
weather_description_transformed = weather_description_transformed.rename(columns={'datetime':'datetime', 'variable':'city','value':'description'})

# wind_direction_transformed.set_index("wind_direction",inplace=True)
weather_description_transformed.head()
```

### Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/weather"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
try:
    city_transformed.to_sql(name='city', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:
    humidity_transformed.to_sql(name='humidity', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:    
    wind_speed_transformed.to_sql(name='wind_speed', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:   
    wind_direction_transformed.to_sql(name='wind_direction', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:    
    temp_df.to_sql(name='temperature', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:    
    pressure_transformed.to_sql(name='pressure', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```

```python
try:    
    weather_description_transformed.to_sql(name='description', con=engine, if_exists='append', index=False)
except :
    print("Duplicate Data")
```
